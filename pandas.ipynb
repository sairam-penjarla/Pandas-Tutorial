```python
import pandas as pd

# Sample DataFrames
df1 = pd.DataFrame({'key': ['A', 'B', 'C', 'D'],
                    'value': [1, 2, 3, 4]})
df2 = pd.DataFrame({'key': ['B', 'D', 'E', 'F'],
                    'value': [5, 6, 7, 8]})
display(df1)
display(df2)
```

```python
left_merge = pd.merge(df1, df2, on='key', how='left')
display(left_merge)
```

```python
right_merge = pd.merge(df1, df2, on='key', how='right')
display(right_merge)
```

```python
inner_merge = pd.merge(df1, df2, on='key', how='inner')
display(inner_merge)
```

```python
outer_merge = pd.merge(df1, df2, on='key', how='outer')
display(outer_merge)
```

```python
df3 = pd.DataFrame({
    'key_left': ['A', 'B', 'C', 'D'],
    'value_left': [1, 2, 3, 4]
})
df4 = pd.DataFrame({
    'key_right': ['B', 'D', 'E', 'F'],
    'value_right': [5, 6, 7, 8]
})
display(df3)
display(df4)
```

```python
left_on_right_on_merge = pd.merge(df3, df4, left_on='key_left', right_on='key_right', how='inner')
display(left_on_right_on_merge)
```

```python
suffix_merge = pd.merge(df1, df2, on='key', how='outer', suffixes=('_left', '_right'))
display(suffix_merge)
```

```python
indicator_merge = pd.merge(df1, df2, on='key', how='outer', indicator=True)
display(indicator_merge)
```

```python
df = pd.DataFrame({
    'Team': ['A', 'B', 'A', 'B', 'A', 'B'],
    'Player': ['Player1', 'Player2', 'Player3', 'Player4', 'Player5', 'Player6'],
    'Points': [20, 25, 15, 30, 18, 22]
})
display(df)
```

```python
grouped = df.groupby('Team')

sum_points = grouped['Points'].sum() # .mean() .max() .min()
display(sum_points)
```

```python
agg_results = grouped['Points'].agg(['sum', 'mean', 'max'])
display(agg_results)
```

```python
agg_diff_cols = grouped.agg({'Points': 'sum', 'Player': 'count'})
display(agg_diff_cols)
```
